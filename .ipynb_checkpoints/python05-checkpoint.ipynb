```python
# í…ŒìŠ¤íŠ¸ 9 ì€í–‰ì˜ˆê¸ˆ ìº¡ìŠí™” ì˜ˆì œ, í´ë˜ìŠ¤ ë³€ìˆ˜
class BankAccount:
    interest_rate = 0.02  # í´ë˜ìŠ¤í¸ìˆ˜
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance or 0 # ìº¡ìŠí™” í•´ì£¼ë©´ ë°–ì—ì„œ ì ìš©ì´ ì•ˆë¨
    def bprint(self):
        return f"{self.owner}ë‹˜ì˜ ì”ê³ ëŠ” {self.__balance}ì› ì…ë‹ˆë‹¤."
    def deposit(self,amount):
        self.__balance += amount
        return f"{self.owner}ë‹˜ì˜ ê³„ì¢Œì— {amount}ì›ì´ ì…ê¸ˆ ë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì”ì•¡: {self.__balance}ì›"
    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance -= amount
            return f"{self.owner}ë‹˜ì˜ ê³„ì¢Œì— {amout}ì›ì´ ì¶œê¸ˆ ë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì”ì•¡ : {self.__balance}ì›"
        return "ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤." # if ì¡°ê±´ì´ ê±°ì§“ì´ë©´ ì‹¤í–‰ë˜ì§€ ì•Šê¸° ë•Œë¬¸ì— elseë‘ ê°™ì€ ì—´ì— ë°”ë¡œ returnì¤˜ë„ ë¨
    def interest(self):
        total = self.__balance * (1 + self.interest_rate)
        return f"({self.interest_rate}%ì˜ ì´ìê°€ ë°˜ì˜í•˜ëŠ” ê¸ˆì•¡ì€ {total:0f}ì› ì…ë‹ˆë‹¤."

# ì¸ìŠ¤í„´ìŠ¤í™”
acc3 = BankAccount("ì² í¬", 50000)
acc4 = BankAccount("ì˜í¬", 200000)
print(acc3.bprint())
print(acc4.bprint())
print(acc3.interest())
print(acc4.interest())

# í´ë˜ìŠ¤ ë³€ìˆ˜ ë°”ê¾¸ê¸°
BankAccount.interest_rate = 0.03
print(acc3.interest())
print(acc4.interest())

# ê³µê²©í›„
# acc3.balance = 9000000
# print(acc3.balance)

# ë‚´ì¥ë©”ì†Œë“œë¡œ ìº¡ìŠí™”ëœ ë³€ìˆ˜ ì ‘ê·¼
# print(acc3.deposit(30000))
# print(acc3.withdraw(5500))
```

```python
class VendingMachine:
    def __init__(self):
        """ìë™íŒë§¤ê¸° ì´ˆê¸°í™”"""
        self.balance = 0 # ì‚¬ìš©ìì˜ ì”ì•¡
        self.items = {
            "ì½œë¼": 1500,
            "ì‚¬ì´ë‹¤":1400,
            "ë¬¼": 1000
        }
    def insert_money(self, amount):
        """ì”ì•¡ ì¶©ì „"""
        self.balance += amount
        return f"ì”ì•¡ì´ {self.balance}ì›ìœ¼ë¡œ ì¦ê°€í–ˆìŠµë‹ˆë‹¤."
    def buy_item(self,item_name):
        """ìƒí’ˆ êµ¬ë§¤"""
        if item_name not in self.items:
            return f"{item_name}ì€(ëŠ”) íŒë§¤í•˜ì§€ ì•ŠëŠ” ìƒí’ˆì…ë‹ˆë‹¤."
        price = self.items[item_name]
        if self.balance < price:
            return f"ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤! ({item_name} ê°€ê²©: {price}ì›, í˜„ì¬ ì”ì•¡: {self.balance}ì›)"
        self.balance -= price:
        return f"{item_name}ì„(ë¥¼) êµ¬ë§¤í–ˆìŠµë‹ˆë‹¤! ë‚¨ì€ ì”ì•¡: {self.balance}ì›"
    def refund(self):
        """ì”ì•¡ ë°˜í™˜"""
        amount = self.balance
        self.balance = 0
        return f"ì”ì•¡ {amount}ì›ì´ ë°˜í™˜ë˜ì—ˆìŠµë‹ˆë‹¤."

# ìë™íŒë§¤ê¸° ê°ì²´ ìƒì„±
vm = VendingMachine()
#ì”ì•¡ ì¶©ì „
print(vm.insert_money(2000)) # ì”ì•¡ì´ 2000ì›ìœ¼ë¡œ ì¦ê°€í–ˆìŠµë‹ˆë‹¤.
#ìƒí’ˆ êµ¬ë§¤
print(vm.buy_item("ì½œë¼")) # ì½œë¼ì„(ë¥¼) êµ¬ë§¤í–ˆìŠµë‹ˆë‹¤! ë‚¨ì€ ì”ì•¡ : 500ì›
print(vm.buy_item("ì‚¬ì´ë‹¤"))
#ì”ì•¡ ë°˜í™˜
#print(vm.buy_item)
```

```python
import random
class Character:
    def __init__(self, name, hp, attack_p):
        """ìºë¦­í„° ì´ˆê¸°í™”"""
        self.name = name
        self.hp = hp
        self.attack_p = attack_p
    def attack(self, opponent):
        """ìƒëŒ€ë°©ì„ ê³µê²©"""
        damage = random.randint(2, self.attack_p)
        opponent.hp -= damage
        print(f"{self.name}ê°€ {opponent.name}ë¥¼ ê³µê²©í•˜ì—¬ {damage} ë°ë¯¸ì§€ë¥¼ ì…í˜”ìŠµë‹ˆë‹¤.")
    def now_status(self):
        """í˜„ì¬ ì²´ë ¥ ìƒíƒœ í™•ì¸"""
        return f"{self.name} ì²´ë ¥:{self.hp}"
    def is_alive(self):
        """ìºë¦­í„° ìƒì¡´ ì—¬ë¶€ í™•ì¸"""
        return self.hp > 0 # ë¹„êµ ì—°ì‚°ìë¥¼ ë„£ì–´ì£¼ë©´ ì°¸ ê±°ì§“ ë°˜í™˜í•´ì¤Œ
```

```python
# íˆì–´ë¡œì™€ ëª¬ìŠ¤í„° ìƒì„±
player = Character("ì˜ì›…", hp=50, attack_p=15) # nameì€ ìœ„ì¹˜ëŠ” í•„ìˆ˜ì˜ì—­ì´ë¼ 
monster = Character("ê³¨ë ˜", 60, 9)
```

```python
# ì „íˆ¬ ì§„í–‰
print("====ì „íˆ¬ì‹œì‘!!====")
while player.is_alive() and monster.is_alive(): # í•œëª… ì£½ìœ¼ë©´ ê²Œì„ ì¤‘ë‹¨
    input("ê³µê²©í•˜ë ¤ë©´ ì—”í„°ë¥¼ ì¹˜ì„¸ìš”.")
    player.attack_p(monster)
    if monster.is_alive():
        monster.attack_p(player)
    print(player.now_status(),' vs ', monster.now_status())
# ê²°ê³¼
if player.is_alive():
    print("ğŸ‰ë‹¹ì‹ ì´ ìŠ¹ë¦¬í•˜ì˜€ìŠµë‹ˆë‹¤.ğŸ‰")
else:
    print("ğŸ©»íŒ¨ë°°í•˜ì˜€ìŠµë‹ˆë‹¤...ğŸ©» to be continue")
```
