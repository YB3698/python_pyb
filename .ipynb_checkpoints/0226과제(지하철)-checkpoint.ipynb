```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import squarify

file_path = "./ref/서울시지하철호선별역별승하차인원정보.csv"
df = pd.read_csv(file_path,encoding="EUC-KR")
# data3 = df.groupby(['호선명'])
data = df[df['호선명']=='3호선']
total_data = data[['역명','승차총승객수']]

d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
plt.rcParams["font.family"] = "D2Coding"

plt.figure(figsize=(20,10))
squarify.plot(sizes=total_data['승차총승객수'],label=[f"{x[0]}({x[1]}명)" for x in zip(total_data['역명'],total_data['승차총승객수'])],alpha=0.7)
plt.rcParams['font.size'] = 10 
# plt.rcParams['text.color']='blue'
# plt
plt.gca().invert_yaxis() # 축 뒤집기
plt.axis("off") # 축 인덱스 없애기
plt.show()
```
