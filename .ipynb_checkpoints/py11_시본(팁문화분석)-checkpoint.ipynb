```python
import seaborn as sns
sns.get_dataset_names()
```

anagrams: 단어 퍼즐과 관련된 데이터셋으로, 실험 데이터를 포함합니다.

anscombe: Anscombe's quartet 데이터셋으로, 서로 다른 데이터 분포가 동일한 통계치를 가질 수 있다는 것을 보여줍니다.

attention: 주의 집중과 관련된 데이터를 담고 있습니다.

brain_networks: 뇌 네트워크와 연결성을 보여주는 데이터입니다.

car_crashes: 자동차 사고 데이터로, 사고율과 연관된 요인을 분석할 수 있습니다.

diamonds: 다이아몬드의 가격, 무게, 색상 등에 대한 데이터를 담고 있습니다.

dots: 시각적 시간적 효과를 나타내는 실험 데이터입니다.

dowjones: 다우존스 지수와 관련된 금융 데이터입니다.

exercise: 운동 유형과 관련된 데이터를 포함합니다.

flights: 항공 승객 수의 월별 변화를 보여주는 데이터입니다.

fmri: 뇌 활동 데이터를 나타내며 fMRI 결과를 포함합니다.

geyser: Old Faithful 간헐천의 분출 패턴 데이터를 담고 있습니다.

glue: GLUE 벤치마크 데이터를 기반으로 텍스트 분류 작업을 지원합니다.

healthexp: 국가별 건강 지출 데이터를 제공합니다.

iris: 붓꽃 종에 따른 꽃잎과 꽃받침의 치수 데이터입니다.

mpg: 자동차의 연비 데이터로, 연료 효율성을 분석할 수 있습니다.

penguins: 펭귄 종과 신체 치수를 포함하는 생태학 데이터입니다.

planets: 외계 행성 데이터로, 천문학 연구에 유용합니다.

seaice: 해빙(바다 얼음) 데이터를 분석할 수 있는 자료입니다.

taxis: 뉴욕 택시 운행과 관련된 데이터를 담고 있습니다.

tips: 레스토랑 팁 데이터로, 팁 문화 분석 연습에 유용합니다.

titanic: 타이타닉 생존자 데이터로, 분류 모델링에 자주 사용됩니다.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x='day', y='tip', hue='sex', data=tips, alpha = 0.7, ax=ax1)
sns.swarmplot(x='day',y='tip',hue='sex',data=tips, palette='Set2',alpha=0.7,ax=ax2)
ax1.set_title('Strip plot 으로 표현한 날짜별로 분류한 성별과 팁 관계')
ax2.set_title('Swarm plot 으로 표현한 날짜별로 분류한 성별과 팁 관계')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.countplot(x='time', data=tips, ax=ax1)
sns.countplot(x='day',hue='day', data=tips, palette='Set2',ax=ax2)
ax1.set_title('Countplt 시간에 따른 팁 빈도차트')
ax2.set_title('Countplt 시간과 날짜에 따른 팁 빈도차트')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x="total_bill",y="tip", data=tips, color="blue", scatter_kws={"s":50,"alpha":0.5}, line_kws={"linestyle":"--"},ax=ax1)
sns.regplot(x="total_bill",y="tip", data=tips, color="blue", scatter_kws={"s":50,"alpha":0.5}, fit_reg=False, ax=ax2)

fig.suptitle("Scatter plots")
ax1.set_title("선형회귀 ON")
ax2.set_title("선형회귀 OFF")
plt.show()
```
