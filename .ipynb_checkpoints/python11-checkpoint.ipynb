# 판다스(pandas) 심화

```python
import numpy as np
import pandas as pd
print("DataFrame: ")
data = {
    'col1':[2,1,9,8,7,4],
    'col2':['A','A','B',np.nan,'D','C'],
    'col3':[0,1,9,4,2,3]
}
index = [1,3,2,5,0,4]
df = pd.DataFrame(data,index=index)
df
```

```python
df = df.sort_index(axis=1, ascending=False)
df
```

```python
# sort_index axis = 0(행) 을 축으로 정렬
df = df.sort_index(axis=0)
df
```

```python
# sort_value 1
sorted_df1 = df.sort_values('col1',ascending=True)
sorted_df1
```

```python
# sort_value 2
sorted_df2 = df.sort_values(['col2','col1'],ascending=[True,False])
sorted_df2
```

```python
x = np.random.rand(10,5)
x
```

```python
x = np.random.normal(5,3,(10,5)) # 평균이, 표준편차 3
print(x)
print(np.mean(x),np.std(x),x.min(),x.max())
```

```python
print(np.percentile(x,0)) # 최소값
print(np.percentile(x,25)) # 
print(np.percentile(x,50)) # 
print(np.percentile(x,75)) # 
print(np.percentile(x,)) # 
```

```python
# 데이터 프레임의 
import numpy as np
import pandas as pd
data = {
    'korean':[50,60,70,80],
    'math':[30,np.nan,40,20]
}
index = ['a','b','c','d']
df = pd.DataFrame(data, index=index)
print(df.index)
print(df.columns)
df
```

```python
print(df.count(axis=0)) # axis = 0 기본값(생략가능)
print(df.count(axis=1))
```

```python
# max(), min(), sum(), mean()
print(df.max())
print(df.max(axis=1))
print(df.min())
print(df.min(axis=1))
print(df.sum())
print(df.sum(axis=1))
print(df.mean())
print(df.mean(axis=1,skipna=False)) # NaN값 보이게 하기
```

```python
# 결측값 대체 fillna
avg = df['math'].mean()
df['math']=df['math'].fillna(avg)
df
```

## 1.3. 그룹으로 묶기(groupby)

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'key' : ['A','B','C','A','B','C'],
    'data1':[1,2,3,1,2,3],
    'data2':[4,4,6,0,6,1],
    # 'data3':['안녕','a','b','하세요','pple','3']
})
print('DataFrame:')
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby(['key','data1']).sum() #.reset_index()
```

```python
df.groupby('key').aggregate(['min','median','max'])
```

```python
df.groupby('key').aggregate({'data1':'max','data2':'sum'})
```
