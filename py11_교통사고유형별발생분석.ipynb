```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# import matplotlib.dates as mdates
import pandas as pd
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
matplotlib.rcParams["font.family"]="D2Coding"
```

```python
file_path = "./ref/.csv"
ori = pd.read_csv(file_path,encoding="EUC-KR")
ori
```

```python
ori_data = ori.groupby('')[''].sum().reset_index(name="count")
ori_data = ori_data.sort_values('count',ascending=False)
ori_data = ori_data.rename(columns={'':'type'})
ori_data
```

```python
ori_data.plot(
    kind = 'pie', 
    y='count',
    labels=ori_data['type'], 
    figsize=(7,7), 
    startangle=90, 
    autopct='%2.1f%%',
    title='    ', 
    cmap=matplotlib.cm.Blues
)
plt.savefig('./chart/accideng.png')
plt.show()
```

```python
ex_data = ori.copy()
ex2_data = ex_data['']
ex2_data
```

```python
ex_data = ori.copy()
ex_data = ex_data.rename(columns={'':'year','':'type','':'accident'})
ex_data
arr_data = list(set(ex_data['type']))
data_list=[]
for t in arr_data:
    data_list.append(ex_data[ex_data['type']==t])
print(arr_data,data_list)
```

```python
total_list = []

for d in data_list:
    result = d.groupby('year')["accident"].sum().reset_index(name='count')
    total_list.append(result)

# print(arr_data,total_list)

for i in range(len(arr_data)):
    total_list[i].plot(
        kind = 'pie',
        y = 'count',
        labels = total_list[i]['year'],
        figsize=(7,7),
        startangle=90,
        autopct = '%.1f%%',
        title = f' {arr_data[i]}   ',
        cmap = matplotlib.cm.Blues)
    plt.savefig(f"./chart/chart{i}.png")
    plt.show()
```
