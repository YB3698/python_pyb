# 서울 CCTV 연계 현황

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm
plt.rc
file_path = "./ref/서울시안심이CCTV연계현황.csv"
CCTV = pd.read_csv(file_path,encoding="CP949")
cctv = CCTV.drop(columns=["안심 주소","위도","경도","수정 일시"])
cctv.columns = ["local","count"]
cctv = cctv.groupby('local').sum()
cctv
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
plt.rcParams["font.family"] = "D2Coding"
cctv['count'].sort_values().plot(kind="barh",figsize=(7,7))
plt.title("서울 CCTV 연계 현황")
plt.show()
```
