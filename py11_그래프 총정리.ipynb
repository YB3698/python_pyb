# 데이터 시각화(Matplotlib)

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.show
```

```python
import numpy as np
x = np.arange(0,2*np.pi,0.01)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.bar([1,2,3,4],[10,20,30,40], width=0.5) # 넓이 지
plt.show()
```

```python
plt.barh(range(1,8,2),[10,20,30,40],height=0.5)
plt.barh(range(2,9,2),[20,15,50,40],height=0.5)
plt.show()
```

```python
plt.hist([1,1,2,3,4,5,6,7,8,10],bins=10)
plt.show()
```

```python
plt.scatter(range(1,11),[30,12,15,14,44,50,60,80,10,15])
plt.show()
```

```python
x = np.random.randint(-100,100,100)
y = np.random.randint(-200,200,100)
z = np.random.randint(0,200,100)
plt.scatter(x,y,s=z*1,alpha=0.5) # alpha는 투명도
plt.show()
```

```python
# np.random.randint(low,high,size) low부터 high까지 랜덤하게 size개의 정수 생성
x = np.random.randint(-100,100,100)
y = np.random.randint(-200,200,100)
z = np.random.randint(0,200,100) # 0~200까지 100만큼 step뛰어서 랜덤지정
plt.scatter(x,y,s=z*1,c=z, alpha=0.5) # s는 사이즈, c는 컬런데 저기에 z값을 넣으며 둘 다 랜덤으로 지정됨
plt.colorbar() # 색 값 표시
plt.show()
```

```python
# pie 그래프 그리기
plt.pie([20,80,10])
plt.show()
```

```python
# pie 그래프 90도 회전
plt.pie([20,80,10], startangle=90)
plt.show()
```

```python
size = [2241,2312,1031,1233]
label = ['A형', 'B형', 'AB형', 'O형']
color = ['darkmagenta','deeppink','hotpink','pink']
plt.axis('equal')
plt.pie(size, labels = label, autopct = '%.1f%%', explode = (0,0,0.1,0), colors = color)
plt.legend()
plt.show()
```
