```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm
file_path = "./ref/_20231231.csv"
CCTV = pd.read_csv(file_path,encoding="EUC-KR")
cctv = CCTV.groupby('').sum() 
cctv = cctv.drop(columns=[""]) #   
resion = cctv.loc[:,cctv.columns.str.contains('|||||')]
resion.columns = resion.columns.str[:2]
resion = resion.T.groupby(lambda x:x[:2]).sum().T
resion
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
plt.rcParams["font.family"] = "D2Coding"

fig, ax = plt.subplots(2,3,figsize=(10,8))
resion[''].sort_values().plot(kind="barh",ax=ax[0][0])
ax[0][0].set_title("  ")
resion[''].sort_values().plot(kind="barh",ax=ax[0][1])
ax[0][1].set_title("  ")
resion[''].sort_values().plot(kind="barh",ax=ax[0][2])
ax[0][2].set_title("  ")
resion[''].sort_values().plot(kind="barh",ax=ax[1][0])
ax[1][0].set_title("  ")
resion[''].sort_values().plot(kind="barh",ax=ax[1][1])
ax[1][1].set_title("  ")
resion[''].sort_values().plot(kind="barh",ax=ax[1][2])
ax[1][2].set_title("  ")
plt.show()
```
