```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm
file_path = "./ref/경찰청범죄발생지역별통계_20231231.csv"
CCTV = pd.read_csv(file_path,encoding="EUC-KR")
cctv = CCTV.groupby('범죄대분류').sum() 
cctv = cctv.drop(columns=["범죄중분류"]) # 필요없는 열 삭제
resion = cctv.loc[:,cctv.columns.str.contains('서울|경남|경북|전남|전북|제주')]
resion.columns = resion.columns.str[:2]
resion = resion.T.groupby(lambda x:x[:2]).sum().T
resion
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
plt.rcParams["font.family"] = "D2Coding"

fig, ax = plt.subplots(2,3,figsize=(10,8))
resion['서울'].sort_values().plot(kind="barh",ax=ax[0][0])
ax[0][0].set_title("서울 범죄 현황")
resion['경남'].sort_values().plot(kind="barh",ax=ax[0][1])
ax[0][1].set_title("경남 범죄 현황")
resion['경북'].sort_values().plot(kind="barh",ax=ax[0][2])
ax[0][2].set_title("경북 범죄 현황")
resion['전남'].sort_values().plot(kind="barh",ax=ax[1][0])
ax[1][0].set_title("전남 범죄 현황")
resion['전북'].sort_values().plot(kind="barh",ax=ax[1][1])
ax[1][1].set_title("전북 범죄 현황")
resion['제주'].sort_values().plot(kind="barh",ax=ax[1][2])
ax[1][2].set_title("제주 범죄 현황")
plt.show()
```
