```python
import pandas as pd
file_path = "./ref/청소년+정신건강_20250304110751.csv"
ori = pd.read_csv(file_path)
ori

# 컬럼명 지정
new_name = ['시점','스트레스','스트레스남학생','스트레스여학생','우울감경험율','우울남학생', '우울여학생','자살생각율','자살남학생','자살여학생']
ori.columns= new_name

#시점을 없애고 , 0번 행도 없앰.
ori= ori.loc[1:,"스트레스":]
# 2번 인덱스 행을 생성 , 100에서 값을 뺀 값
ori.loc[3] = ori.loc[2]
ori.loc[2] = 100 - ori.loc[1,'스트레스':].astype(float)
ori.loc[4] = 100 - ori.loc[3,'스트레스':].astype(float)

# 마지막에 응답컬럼을 생성, 그렇다 or 아니다.
ori['응답'] = ['2023년 기준 그렇다','2023년 기준 아니다','2024년 기준 그렇다','2024년 기준 아니다']
ori = ori.set_index('응답')
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm
# 윈도우용 폰트 경로 설정
path = "C:\\Windows\\Fonts\\NanumGothic.ttf"
font_name = fm.FontProperties(fname=path).get_name()
rc('font',family=font_name)

fig , ax = plt.subplots(2,3,figsize=(10,8)) 

ori.loc[['2023년 기준 그렇다','2023년 기준 아니다'],'스트레스'].astype(float).plot.pie(explode = [0, 0.1], ax = ax[0][0], autopct="%.1f%%")
# 계산을 위해 타입을 float로 바꿔주고, explode를 이용해 파이 그래프 조각 간격을 띄워 보기 좋게 해주고, 해당 자료를 어디 그래프에 넣을지 위치 설정해주기
ax[0][0].set_title('스트레스를 받은적 있다.')
ax[0][0].set_ylabel('2023년')

ori.loc[['2023년 기준 그렇다','2023년 기준 아니다'],'우울감경험율'].astype(float).plot.pie(explode = [0, 0.1], ax = ax[0][1], autopct="%.1f%%")
# 계산을 위해 타입을 float로 바꿔주고, explode를 이용해 파이 그래프 조각 간격을 띄워 보기 좋게 해주고, 해당 자료를 어디 그래프에 넣을지 위치 설정해주기
ax[0][1].set_title('우울증을 경험한적 있다.')
ax[0][1].set_ylabel('2023년')

ori.loc[['2023년 기준 그렇다','2023년 기준 아니다'],'자살생각율'].astype(float).plot.pie(explode = [0, 0.1], ax = ax[0][2], autopct="%.1f%%")
# 계산을 위해 타입을 float로 바꿔주고, explode를 이용해 파이 그래프 조각 간격을 띄워 보기 좋게 해주고, 해당 자료를 어디 그래프에 넣을지 위치 설정해주기
ax[0][2].set_title('자살을 고민한 적 있다.')
ax[0][2].set_ylabel('2023년')

ori.loc[['2024년 기준 그렇다','2024년 기준 아니다'],'스트레스'].astype(float).plot.pie(explode = [0, 0.1], ax = ax[1][0], autopct="%.1f%%")
# 계산을 위해 타입을 float로 바꿔주고, explode를 이용해 파이 그래프 조각 간격을 띄워 보기 좋게 해주고, 해당 자료를 어디 그래프에 넣을지 위치 설정해주기
ax[1][0].set_title('스트레스를 받은적 있다.')
ax[1][0].set_ylabel('2024년')

ori.loc[['2024년 기준 그렇다','2024년 기준 아니다'],'우울감경험율'].astype(float).plot.pie(explode = [0, 0.1], ax = ax[1][1], autopct="%.1f%%")
# 계산을 위해 타입을 float로 바꿔주고, explode를 이용해 파이 그래프 조각 간격을 띄워 보기 좋게 해주고, 해당 자료를 어디 그래프에 넣을지 위치 설정해주기
ax[1][1].set_title('우울증을 경험한적 있다.')
ax[1][1].set_ylabel('2024년')

ori.loc[['2024년 기준 그렇다','2024년 기준 아니다'],'자살생각율'].astype(float).plot.pie(explode = [0, 0.1], ax = ax[1][2], autopct="%.1f%%")
# 계산을 위해 타입을 float로 바꿔주고, explode를 이용해 파이 그래프 조각 간격을 띄워 보기 좋게 해주고, 해당 자료를 어디 그래프에 넣을지 위치 설정해주기
ax[1][2].set_title('자살을 고민한 적 있다.')
ax[1][2].set_ylabel('2024년')

plt.suptitle('2023 ~ 2024년 청소년 정신건강관련 통계',fontsize=16)
plt.show()
```
