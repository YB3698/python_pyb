```python
import pandas as pd
file_path = "./ref/+_20250304110751.csv"
ori = pd.read_csv(file_path)
ori

#  
new_name = ['','','','','','', '','','','']
ori.columns= new_name

#  , 0  .
ori= ori.loc[1:,"":]
# 2    , 100   
ori.loc[3] = ori.loc[2]
ori.loc[2] = 100 - ori.loc[1,'':].astype(float)
ori.loc[4] = 100 - ori.loc[3,'':].astype(float)

#   ,  or .
ori[''] = ['2023  ','2023  ','2024  ','2024  ']
ori = ori.set_index('')
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm
#    
path = "C:\\Windows\\Fonts\\NanumGothic.ttf"
font_name = fm.FontProperties(fname=path).get_name()
rc('font',family=font_name)

fig , ax = plt.subplots(2,3,figsize=(10,8)) 

ori.loc[['2023  ','2023  '],''].astype(float).plot.pie(explode = [0, 0.1], ax = ax[0][0], autopct="%.1f%%")
#    float , explode         ,       
ax[0][0].set_title('  .')
ax[0][0].set_ylabel('2023')

ori.loc[['2023  ','2023  '],''].astype(float).plot.pie(explode = [0, 0.1], ax = ax[0][1], autopct="%.1f%%")
#    float , explode         ,       
ax[0][1].set_title('  .')
ax[0][1].set_ylabel('2023')

ori.loc[['2023  ','2023  '],''].astype(float).plot.pie(explode = [0, 0.1], ax = ax[0][2], autopct="%.1f%%")
#    float , explode         ,       
ax[0][2].set_title('   .')
ax[0][2].set_ylabel('2023')

ori.loc[['2024  ','2024  '],''].astype(float).plot.pie(explode = [0, 0.1], ax = ax[1][0], autopct="%.1f%%")
#    float , explode         ,       
ax[1][0].set_title('  .')
ax[1][0].set_ylabel('2024')

ori.loc[['2024  ','2024  '],''].astype(float).plot.pie(explode = [0, 0.1], ax = ax[1][1], autopct="%.1f%%")
#    float , explode         ,       
ax[1][1].set_title('  .')
ax[1][1].set_ylabel('2024')

ori.loc[['2024  ','2024  '],''].astype(float).plot.pie(explode = [0, 0.1], ax = ax[1][2], autopct="%.1f%%")
#    float , explode         ,       
ax[1][2].set_title('   .')
ax[1][2].set_ylabel('2024')

plt.suptitle('2023 ~ 2024   ',fontsize=16)
plt.show()
```
