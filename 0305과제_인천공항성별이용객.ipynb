```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

file_path = "./ref/20250305.csv"
ori = pd.read_csv(file_path, encoding="EUC-KR")
ex1_data = ori.copy()
ex1_data.columns = ["Airport","type1","gender","year","type2","value"]

man = ex1_data['gender']==""
ex_man = ex1_data[man]
woman = ex1_data['gender']==""
ex_woman = ex1_data[woman]
ex2_data = ex_man.groupby('year')['value'].sum().reset_index(name="total")
ex3_data = ex_woman.groupby('year')['value'].sum().reset_index(name="total")

from matplotlib import rc
path = "C:\\Windows\\Fonts\\NanumGothic.ttf"
font_name = fm.FontProperties(fname=path).get_name()
rc('font',family=font_name)

fig, axs = plt.subplots(1,3,figsize=(16,8))
axs[0].plot(ex2_data,color="b")
axs[1].plot(ex3_data,color="r")
axs[2].plot(ex2_data,color="b")
axs[2].plot(ex3_data,color="r")
axs[0].set_title("  ")
axs[1].set_title("  ")
axs[2].set_title("   ")

plt.tight_layout()
plt.show()
```
