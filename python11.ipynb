```python
# 데이터 프레임의 
import numpy as np
import pandas as pd
data = {
    'korean':[50,60,70,80],
    'math':[30,np.nan,40,20]
}
index = ['a','b','c','d']
df = pd.DataFrame(data, index=index)
print(df.index)
print(df.columns)
df
```

```python
print(df.count(axis=0)) # axis = 0 기본값(생략가능)
print(df.count(axis=1))
```

```python
# max(), min(), sum(), mean()
print(df.max())
print(df.max(axis=1))
print(df.min())
print(df.min(axis=1))
print(df.sum())
print(df.sum(axis=1))
print(df.mean())
print(df.mean(axis=1,skipna=False)) # NaN값 보이게 하기
```

```python
# 결측값 대체 fillna
avg = df['math'].mean()
df['math']=df['math'].fillna(avg)
df
```

## 1.3. 그룹으로 묶기(groupby)

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'key' : ['A','B','C','A','B','C'],
    'data1':[1,2,3,1,2,3],
    'data2':[4,4,6,0,6,1],
    # 'data3':['안녕','a','b','하세요','pple','3']
})
print('DataFrame:')
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby(['key','data1']).sum() #.reset_index()
```

```python
df.groupby('key').aggregate(['min','median','max'])
```

```python
df.groupby('key').aggregate({'data1':'max','data2':'sum'})
```

```python
# filter 메서드
t = list(range(1,11))
result = []
def is_even(x):
    return True if x%2==0 else False
    # if를 기준으로 참이면 왼쪽, 거짓이면 오른쪽 결과가 나옴
    # 다른 언어에선 if조건 ? : 실행1, 실행2 로 하는데 파이썬만 이렇게 사용함
for i in t:
    if is_even(i):
        result.append(i)
print(result)
```

```python
df.groupby('key').mean()
```

```python
def filter_mean(x):
    return x ['data2'].mean() > 3
df.groupby('key').filter(filter_mean)
```

```python
df1 = df.groupby('key').get_group('A')
df2 = df.groupby('key').get_group('B')
df3 = df.groupby('key').get_group('C')
df4 = pd.concat([df1,df3],axis=1)
df5 = pd.concat([df2,df3],axis=1)
# 평균값으로 채운다.


# means = df2['data2]'.astype(int).mean()
# # means
# df5.fillna(1)
```

```python
import pandas as pd
df = pd.DataFrame([
    ['20210901', 'A',   1],
    ['20210901', 'B',  10],
    ['20210901', 'C', 100],
    ['20210902', 'A' ,   2],
    ['20210902', 'B',  20],
    ['20210902', 'C', 200],
    ['20210903', 'A',   3],
    ['20210903', 'B',  30],
    ['20210903', 'C', 300],
    
], columns = ['date','type', 'volume'])
df
```

```python
pivot1 = df.pivot(index="date",columns="type",values="volume")
pivot1
```

```python
pivot1.columns = pivot1.columns.values
pivot1
```

```python
pivot1.reset_index(level=0,inplace=True)
pivot1
```

```python
pivot1 = pivot1.set_index('date')
pivot1
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2,figsize=(12,5))
pivot1.plot(kind='bar',ax=ax[0])
pivot1.plot(kind='barh',ax=ax[1])
plt.show
```

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    '지역':['안동','안동','안동','경기','경기','부산','서울','안동'],
    '요일':['월','월','수','월','화','월','목','금'],
    '강수확률':[80,70,90,10,20,30,40,80],
    '강수량':[80,50,100,200,200,100,50,1000]})
df
```

```python
pd.pivot_table(
    df, 
    index='요일',
    columns='지역',
    aggfunc='max'
).fillna(' ') # fillna는 nan값을 다른 값으로 채울 때 사용
```
